## Anime list data collected plan

Possible columns from "animelist" type data from API

* Anime ID: A unique identifier for each anime in the MyAnimeList database.
* Title: The title of the anime series.
* Status: The user's status for the anime (e.g., "Watching," "Completed," * "On-Hold," "Dropped," "Plan to Watch").
* Score: The user's rating or score for the anime.
* Watched Episodes: The number of episodes the user has watched.
* Total Episodes: The total number of episodes in the anime series.
* Start Date: The date when the user started watching the anime.
* Finish Date: The date when the user finished watching the anime.
* Priority: Priority level set by the user.
* Tags: User-defined tags or labels for the anime.
* Rewatching: Whether the user is rewatching the anime.
* Rewatch Count: The number of times the user has rewatched the anime.
* Rewatch Value: User's rating or score for the rewatch.
* Comments: User comments or notes about the anime.

Goal: Create a dataset on which we can perform collaborative filtering.
* Enable mode: suggestions based on 1 anime
* Enable mode: Suggestions based on user history

Constraints of data:
* Cannot suggest a future season if user hasn't already watched the season previous to it
* Don't reccomend any anime that has a score lower than 6.5
* User must have at least 10 reviewed animes

```python
import requests
import numpy as np
import time
import pdb
# This class is used as the main application for collecting data from the myanimelist API

class Data_collector:

  # Define the API base URL
  BASE_URL = "https://api.myanimelist.net/v2/"
  def __init__(self):
    # get api key
    self.api_key = self.get_api_key()
    # test api key auth
    self.print_user_info()

  def get_api_key(self):
    with open('token.json', "r") as f:
      json_file = json.load(f)
      access_token = json_file['access_token']
    return access_token

# Test if current token is up to date, if not, refresh token. (currently, whenver we run the other file it already refreshes the token)

  def print_user_info(self):
      url = 'https://api.myanimelist.net/v2/users/@me'
      response = requests.get(url, headers = {
          'Authorization': f'Bearer {self.api_key}'
          })
      try:
        response.raise_for_status()
        user = response.json()
        response.close()
        print(f"\n>>> Greetings {user['name']}! <<<")
      except:
        print("Access token not allowed, refreshing token")
        %run refresh_token
        # After refreshing the file token.json, we must update the variable access token
        with open('token.json', "r") as f:
          json_file = json.load(f)
          print(json_file)
          self.api_key = json_file['access_token']
          print(access_token)

        # try again if to see if we actually have the auth
        try:
          response.raise_for_status()
          user = response.json()
          response.close()
          print(f"\n>>> Greetings {user['name']}! <<<")
        except:
          print("Token was not able to be refreshed, please debug")

  # Function to get user anime list with scores
  def get_user_anime_list(self, username):

      # username = "TensaiOji"

      # url = f'https://api.myanimelist.net/v2/users/{username}/animelist'
      # url_2 = f'https://api.myanimelist.net/v2/users/{username}/animelist?fields=anime_list_entries.list_status.num_times_rewatched,score'

      # # Set up the headers with your access token
      # headers = {
      #     'Authorization': f'Bearer {access_token}',
      # }

      # # Send a GET request to retrieve the user's anime list
      # response = requests.get(url_2, headers=headers)



      # Endpoint to retrieve user anime list
      endpoint = f"users/{username}/animelist"

      # Parameters for filtering by score and pagination
      params = {
          "fields": "list_status,anime(id,title,score)",
          "status": "completed", # This will filter the results by only the completed anime
          "sort": "list_score",
          "limit": 300,  # Adjust the limit as needed
          "offset": 0,
      }
      # pdb.set_trace()
      # Initialize an empty list to store results
      user_anime_list = []

      while True:
          # Make the API request
          response = requests.get(
              f"{self.BASE_URL}{endpoint}",
              params=params,
              headers={"Authorization": f"Bearer {self.api_key}"}
          )
          time.sleep(0.4)

          if response.status_code == 200:
              data = response.json()
              user_anime_list.extend(data["data"])
              break
              # Check if there are more pages
              # if "paging" in data and "next" in data["paging"]:
              #     params["offset"] += 300
              #     pdb.set_trace()
              # else:
              #     break
          else:
              print(response.json())
              print(f"Error fetching anime list for {username}: {response.status_code}")
              break

      return user_anime_list

  # Function to filter users with more than 100 reviews and collect their lists
  def collect_user_lists(self, users, min_reviews=100):
      user_lists_with_scores = []

      # runtime metrics
      count = 0
      animelist_collection_times = []

      for user in users:
          start = time.time() # stats
          anime_list = self.get_user_anime_list(user)
          duration = time.time() - start
          animelist_collection_times.append(duration)
          if duration >= 1:
            print(f"Warning: {user}'s animelist took {duration} secs to be collected")

          if count%10 == 0:
            print(f"{count} users animelist collected")

          count += 1

          # Filter users with more than 100 reviews
          if len(anime_list) >= min_reviews:
              user_lists_with_scores.append({"username": user, "anime_list": anime_list})

          time.sleep(0.4)

      print(f"The mean duration was: {np.mean(np.array(animelist_collection_times))}")

      return user_lists_with_scores


# Class End
```

```python
import requests
import json
import math
import traceback
import time

class Jikan:
  # This method requires a page_num argument and provides a list of 10

  def get_users_per_page(page_num):
    base_url = 'https://api.jikan.moe/v4/users'
    page = page_num          # Assuming we want the first page of results
    limit = 20        # Limit the number of results to 10
    # q = 'some_query'  # Some search query
    gender = 'any'    # Accept any gender
    max_age = 50      # Maximum age
    min_age = 18      # Minimum age

    # Construct the full URL with query parameters
    # full_url = f"{base_url}?page={page}&limit={limit}&q={q}&gender={gender}&location={location}&maxAge={max_age}&minAge={min_age}"

    full_url = f"{base_url}?page={page}&limit={limit}&gender={gender}&maxAge={max_age}&minAge={min_age}"

    # Make get request to the API url
    response = requests.get(full_url)
    users = response.json()
    # put response into json format .dumps with indent for visualization
    # print(json.dumps(users, indent = 1)) # for debugging
    response.close()
    try:
      user_list = [item["username"] for item in users["data"]]
    except:
      print(json.dumps(users, indent = 1))
    return user_list

# get_users_per_page(5)

  def get_users(num_users, starting_page_num = 1):
    # Calculate how many pages we need to query
    # We receive 20 users per page so we need to get a number of pages necessary rounded up
    num_pages = math.ceil(num_users/20)

    full_list = []
    for i in range(num_pages):
      try:
        full_list += Jikan.get_users_per_page(starting_page_num + i)
        time.sleep(0.4)
      except Exception as e:
        traceback.print_exc()

    return full_list

temp = Jikan.get_users(50)
```

```python
import os
import pandas as pd
from scipy.sparse import csr_matrix, save_npz, load_npz
import json

def get_batch(min_reviews, num_users, file_path):
    # Get users
    users_to_collect = Jikan.get_users(num_users)
    user_lists_with_scores = dc.collect_user_lists(users_to_collect, min_reviews)
    anime_score_dict = {}

    # Create dataframe of users by anime
    for i, user_data in enumerate(user_lists_with_scores):
        username = user_data["username"]
        anime_score_dict[username] = {node['node']['title']: node['list_status']['score'] for node in user_data['anime_list']}

    # Fill out any missing values from dict with 0s
    df = pd.DataFrame.from_dict(anime_score_dict, orient='index').fillna(0)

    # Convert values to sparse matrix
    sparse_mat = csr_matrix(df.values)
    save_npz(file_path, sparse_mat)

    # Save row labels (usernames) and column labels (anime titles)
    labels_path = file_path.replace('.npz', '_labels.json')
    labels = {
        'rows': df.index.tolist(),  # Usernames
        'columns': df.columns.tolist()  # Anime names
    }

    with open(labels_path, 'w') as f:
        json.dump(labels, f)

    # Return both the DataFrame and the actual number of users processed
    return df, len(user_lists_with_scores)

def create_db(min_reviews, total_users, batch_sizes, folder_path):
    total_users_read = 0
    batch_number = 0

    while total_users_read < total_users:
        current_batch_size = min(batch_sizes, total_users - total_users_read)
        file_path = os.path.join(folder_path, f'batch_{batch_number}.npz')

        # Call your existing function to process this batch and save the sparse matrix
        df, actual_batch_size = get_batch(min_reviews, current_batch_size, file_path)

        total_users_read += actual_batch_size
        batch_number += 1
        print(f'Processed batch {batch_number} with {actual_batch_size} users. Total users processed: {total_users_read}')

    print('Database creation complete.')


def reconstruct_df(matrix_file_path, labels_file_path):
    # Load the sparse matrix
    sparse_matrix = load_npz(matrix_file_path)

    # Load the labels
    with open(labels_file_path, 'r') as f:
        labels = json.load(f)
    row_labels = labels['rows']
    column_labels = labels['columns']

    # Convert the sparse matrix to a DataFrame
    df = pd.DataFrame.sparse.from_spmatrix(sparse_matrix, index=row_labels, columns=column_labels)

    return df
```
